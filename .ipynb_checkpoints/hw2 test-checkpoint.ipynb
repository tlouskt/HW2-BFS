```python
def bfs(G, start, end=None):
    if start not in G:
        return None
    if end and end not in G:
        return None

    visited = []
    q = Queue()
    q.put(start)
    visited.append(start)
    path = []

    while not q.empty():
        v = q.get()
        path.append(v)
        if v == end:
            return path

        neighbors = G.neighbors(v)
        for n in neighbors:
            if n not in visited:
                visited.append(n)
                q.put(n)

    if end:
        return None
    else:
        return path
```

```python
G_test = nx.DiGraph()
G_test.add_edges_from([(0, 1), (1, 2), (1, 3), (1, 4), (3, 5), (3, 6), (4,5), (4,7)])
```

```python
print(bfs(G_test, 0))
print(bfs(G_test, 0, 2)) 
print(bfs(G_test, 4)) 
print(bfs(G_test, 0, 4)) 
```
